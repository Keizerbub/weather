```python
import pandas as pd
df = pd.read_csv('/content/sample_data/tourisme_125 (1).csv')
df.head()
```

```python
def ajout (nom):
  return nom + ' France'

df['Name'] = df['Name'].apply(ajout)

df.head()
```

```python
location_name=list(df['Name'])
```

```python
data = []
```

```python
for location in location_name:
    response = map_client.places(query=location)
    results = response.get('results')
    #locations.append(location)
    data.append({'Name': location, 'results': results})
```

```python
import pandas as pd

# Créer une liste pour stocker les données
df_data = []

for item in data:
    name = item['Name']
    results = item['results']
    for result in results:
        address = result['formatted_address']
        lat = result['geometry']['location']['lat']
        lng = result['geometry']['location']['lng']
        df_data.append({'Nom': name, 'Adresse': address, 'Latitude': lat, 'Longitude': lng})

# Créer un DataFrame à partir de la liste
df = pd.DataFrame(df_data)

print(df)
```

```python
output_file = 'list_tourisme2.csv'

df.to_csv(output_file, index=False)
```
