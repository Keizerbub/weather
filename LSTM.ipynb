```python
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense, LSTM
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
plt.style.use("fivethirtyeight")
%matplotlib inline
```

```python
df=pd.read_csv("/content/1c87202c6695502531c02c9ab79c0e35.csv")
df=df.set_index("dt_iso")
df=df[['temp','visibility','dew_point','feels_like','pressure','humidity','wind_speed']]
df=df.dropna()
df.tail(5)
```

# setting our target column

# Splitting

```python
for column in ['temp','visibility','dew_point','feels_like','pressure','humidity','wind_speed']:
    data=df[column]
    Ntest = int(40000)
    train = data.iloc[:-Ntest]
    test = data.iloc[-Ntest:]


    T = 72 # nombre de valeur à prendre en compte pour faire la prédiction
    X = [] # les variable dont j'ai besoin
    Y = [] # les target

    #changing our dataset in numpy series
    series=data

    # Get the number of rows to train the model on


    for t in range(len(series)-T):
      x = series[t:t+T]
      X.append(x)
      y = series[t+T]
      Y.append(y)

    X = np.array(X).reshape(-1, T)
    Y = np.array(Y)
    N = len(X)
    print("X.shape", X.shape, "Y.shape", Y.shape)

    x_train, y_train = X[:-Ntest], Y[:-Ntest]
    x_test, y_test = X[-Ntest:],Y[-Ntest:]


    # Build the LSTM model
    model = Sequential()
    model.add(LSTM(128, return_sequences=True, input_shape= (x_train.shape[1], 1)))
    model.add(LSTM(64, return_sequences=False))
    model.add(Dense(25))
    model.add(Dense(1))

    # Compile the model
    model.compile(optimizer='adam', loss='mean_squared_error')

    # Train the model
    model.fit(x_train, y_train, batch_size=1000, epochs=5)

    # Save the trained model
    model.save(f"/content/{column}_lstm_model.h5")

    # Get the models predicted price values
    predictions = model.predict(x_test)
    print(f"prediction: {predictions}")

    # Get the root mean squared error (RMSE)
    rmse = np.sqrt(np.mean(((predictions - y_test) ** 2)))
    print(f"RMSE: {rmse}")
```
